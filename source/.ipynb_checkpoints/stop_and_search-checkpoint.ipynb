```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
```

```python
df = pd.read_csv('../files/2020-07/2020-07-cleveland-stop-and-search.csv')
pd.set_option('max_columns', None)
pd.set_option("max_rows", None)
```

```python
mask = df['Age range'].values == '10-17'
df_new = df.loc[mask]
new = df_new.filter(['Age range'],axis=1)
```

```python
df_new.rename(columns = {'Age range':'age_range'}, inplace = True)
```

```python
new.apply(pd.value_counts).plot(kind='bar')
```

```python
request = requests.get("https://data.police.uk/api/stops-force?force=bedfordshire")
```

```python
response = request.json()
```

```python
for item in response:
    print(item['age_range'],": ", item['datetime'])
```
